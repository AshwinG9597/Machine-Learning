# Regression with Moore-Penrose Pseudoinverse

```python
#import the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#lets solve a miniature example of regression with 8 data points
x = [0,1,2,3,4,5,6,7]
y = [1.86,1.31,0.62,0.33,0.09,-0.67,-1.23,-1.37]
```

```python
#scatter plot of all the points
fig,ax = plt.subplots(figsize = (10,5),dpi = 150)
ax.set_title('Clinical Trial')
ax.set_xlabel('Drug Dosage (mL)')
ax.set_ylabel('Forgetfulness')
ax.scatter(x,y)
```

```python
# create a matrix with the features
X = np.concatenate((np.matrix(np.ones(len(x))).T,np.matrix(x).T),axis =1)
X
```

```python
#taking the dot product of the pseudoinverse and the y matrix, we get the weights
w = np.dot(np.linalg.pinv(X),y)
print(w)
y_int = np.ravel(w)[0]
m = np.ravel(w)[1]
print('y intercept is: ',y)
print('slope is: ',m)
```

```python
#plot the line of best fit
fig,ax = plt.subplots(figsize = (10,5),dpi = 150)
ax.set_title('Clinical Trial')
ax.set_xlabel('Drug Dosage (mL)')
ax.set_ylabel('Forgetfulness')
ax.scatter(x,y)
x_min,x_max = ax.get_xlim()
y_min,y_max = m*x_min + y_int,m*x_max + y_int
ax.plot([x_min,x_max],[y_min,y_max],'r')
```

```python
#another way to plot using the data points instead of the x_limits
fig,ax = plt.subplots(figsize = (10,5),dpi = 150)
ax.set_title('Clinical Trial')
ax.set_xlabel('Drug Dosage (mL)')
ax.set_ylabel('Forgetfulness')
ax.scatter(x,y)
ax.plot(x,m*np.array(x) + y_int,'purple')
```
