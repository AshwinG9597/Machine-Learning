# Calculus - The Delta Method

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,10000)
```

```python
def f(my_x):
    y = my_x**2 + 2*my_x + 2
    return y
```

```python
#plot the curve of the given equation between -10,10 x-interval
plt.plot(x,f(x))
plt.scatter(-1,f(-1))
plt.scatter(-2,f(-2))
# plt.plot([-1,-2],[f(-1),f(-2)])
```

```python
#calculate the respective coordinates
x1,y1 = -1,f(-1)
x2,y2 = -2,f(-2)
```

```python
#slope
m = (y2-y1)/(x2-x1)
m
```

```python
#y-intercept calculation
b = y2-m*x2
b
```

```python
#equation of the line
y = m*x + b
```

```python
plt.plot(x,f(x))
plt.scatter(-1,f(-1))
plt.scatter(-2,f(-2))
plt.plot(x,y)
```

As limit approaches its closest value

```python
#calculate respective coordinates for points much closer
x1,y1 = -1,f(-1)
x2,y2 = -1.0001,f(-1.0001)
```

```python
#slope
m = (y2-y1)/(x2-x1)
m
```

```python
#y-intercept calculation
b = y2-m*x2
b
```

```python
#equation of the line
y = m*x + b
```

```python
plt.plot(x,f(x))
plt.scatter(-1,f(-1))
plt.scatter(-1,f(-1.0001))
plt.plot(x,y)
```

Since the lines are so close to each other and are in the local minima, they will produce a slope that equals zero
