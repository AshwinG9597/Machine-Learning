# Eigenvectors and Eigenvalues

Import the necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

An eigenvalue is simply a scalar that scales the eigenvector v such that the following equation is satisfied
<br><br>
Av = Î»v

where A is any matrix and v is the eigenvector

```python
#take a sample matrix A
A = np.array([[-1,4],[2,-2]])
A
```

To find the eigenvalues and eigenvectors of a matrix, we use the numpy linalg library

```python
lambdas, V = np.linalg.eig(A)
```

```python
V
```

The columns of the matrix V are the eigenvectors of the matrix A

To prove the eigenvalue equation for the first vector

```python
lmbda1 = lambdas[0]
v1 = V[:,0]
```

```python
lhs1 = np.dot(A,v1)
rhs1 = lmbda1 * v1
```

```python
print('lhs:',lhs1)
print('rhs:',rhs1)
```

Repeat the same for the second vector

```python
lmbda2 = lambdas[1]
v2 = V[:,1]
```

```python
lhs2 = np.dot(A,v2)
rhs2 = lmbda2 * v2
```

```python
print('lhs:',lhs2)
print('rhs:',rhs2)
```

Multiply the matrix with the eigenvector to get the scaled vector

```python
Av1 = np.dot(A,v1)
Av2 = np.dot(A,v2)
```

```python
def plot_vectors(vectors,colors):
    plt.figure(figsize=(10,7),dpi = 150)
    plt.axhline(0,color = 'lightgray')
    plt.axvline(0,color = 'lightgray')
    plt.xlabel('X axis')
    plt.ylabel('Y axis')
    plt.title('Plotting Vectors')
    for i in range(len(vectors)):
        plt.quiver(0,0,np.real(vectors[i][0]),np.real(vectors[i][1]),angles = 'xy',scale_units = 'xy', 
                   scale = 1, color = colors[i])
```

```python
#plot for the first eigenvector
plot_vectors([Av1,v1],['blue','lightblue'])
plt.xlim(-1,2)
plt.ylim(-1,2)
```

```python
#plot for the second eigenvector
plot_vectors([Av2,v2],['green','lightgreen'])
plt.xlim(-1,5)
plt.ylim(-5,2)
```

Consolidating all the vectors into one graph

```python
plot_vectors([Av1,v1,Av2,v2],['blue','lightblue','green','lightgreen'])
plt.xlim(-2,5)
plt.ylim(-4,2)
```
