## Plotting a system of linear equations

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
t = np.linspace(0,40,1000)
```

Equation for robber (travelling at 150kmph = 2.5 km/min)

```python
d_r = 2.5*t
```

Equation for police (travelling at 180kmph = 3 km/min and starts 5 mins late)

```python
d_p = 3*(t-5)
```

Solving for t

```python
tm = 15/0.5
ds = 2.5*30
print('time taken : ',tm)
print('distance traveled : ',ds)
```

plot the graph where the lines intersect

```python
fig,ax = plt.subplots(dpi = 80, figsize = (6,4))

plt.title('A Bank Robber Caught')
plt.xlabel('time (in minutes)')
plt.ylabel('distance (in kms)')

ax.set_xlim(0,40)
ax.set_ylim(0,100)
plt.plot(t,d_r,color = 'r')
plt.plot(t,d_p,color = 'b')

plt.axhline(y = ds, linestyle = '--')
plt.axvline(x = tm, linestyle = '--')
```
