# Partial Derivatives

Define the function z = x^2 - y^2

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#function definition
def f(my_x,my_y):
    return my_x**2 - my_y**2
```

```python
xs = np.linspace(-3,3,1000)
```

```python
zs_wrt_x = f(xs,0)
```

```python
#plot the function keeping y = 0
plt.axhline(0,color = 'lightgrey')
plt.axvline(0,color = 'lightgrey')
plt.plot(xs,zs_wrt_x)
plt.title('X vs Z')
plt.xlabel('x')
plt.ylabel('z')
```

```python
#determine slope wrt_x => dz/dx
def delz_delx(my_x,my_y):
    return 2*my_x
```

```python
x_vals = [-2,-1,0,1,2]
```

```python
#plot the slopes on the function for the given list of points
plt.figure(figsize=(8, 6))
plt.axhline(0,color = 'lightgrey')
plt.axvline(0,color = 'lightgrey')
plt.plot(xs,zs_wrt_x,color = 'violet',lw = 3)
plt.title('X vs Z')
plt.xlabel('x')
plt.ylabel('z')

for x in x_vals: 
    m = delz_delx(x,0)
    b = f(x,0) - m*x
    plt.scatter(x,f(x,0))
    plt.plot(xs,(m*xs+b),'--')
    plt.xlim(-3,3)
    plt.ylim(-1,10)
```

```python
#plot the function keeping x = 0
plt.axhline(0,color = 'lightgrey')
plt.axvline(0,color = 'lightgrey')
plt.plot(xs,f(0,xs))
plt.title('Y vs Z')
plt.xlabel('y')
plt.ylabel('z')
```

```python
#find slope wrt y
def delz_dely(my_x,my_y):
    return -2*my_y
```

```python
y_vals = [-2,-1,0,1,2]
```

```python
#plot the slopes on the function for the given list of points
plt.figure(figsize=(8, 6))
plt.axhline(0,color = 'lightgrey')
plt.axvline(0,color = 'lightgrey')
plt.plot(xs,f(0,xs),color = 'violet',lw = 3)
plt.title('Y vs Z')
plt.xlabel('y')
plt.ylabel('z')
for y in y_vals:
    m = delz_dely(0,y)
    b = f(0,y) - m*y
    plt.scatter(y,f(0,y))
    plt.plot(xs,(m*xs + b),'--')
plt.xlim(-3,3)
plt.ylim(-10,1)
```
