## Affine Transformations

Import the necessary Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Define a vector x

```python
x = np.array([3,1])
```

Define a function to plot the vectors and their respective colors

```python
def plot_vectors(vectors, colors):
    plt.figure(figsize=(10,7),dpi = 150)
    plt.axhline(0,color = 'lightgray')
    plt.axvline(0,color = 'lightgray')
    plt.xlabel('X axis')
    plt.ylabel('Y axis')
    plt.title('Plotting Vectors')
    for i in range(len(vectors)):
        x = np.concatenate([[0,0],vectors[i]])
        plt.quiver(x[0], x[1], x[2], x[3],angles = 'xy',scale_units = 'xy',scale = 1,color = colors[i],)
```

Plot x on the graph

```python
plot_vectors([x],['lightgrey'])
plt.xlim(-1,5)
plt.ylim(-1,5)
```

Proving that an Identity Matrix multiplied by a vector is equal to the vector itself

```python
I = np.array([[1,0],[0,1]])
Iv = np.dot(I,x)
plot_vectors([x,Iv],['orange','blue'])
plt.xlim(-1,5)
plt.ylim(-1,5) 
```

Flipping a vector about the X-Axis

```python
E = np.array([[1,0],[0,-1]])
Ev = np.dot(E,x)
plot_vectors([x,Ev],['lightblue','blue'])
plt.xlim(-1,5)
plt.ylim(-3,5)
```

Flipping a vector about the Y-Axis

```python
F = np.array([[-1,0],[0,1]])
Fv = np.dot(F,x)
plot_vectors([x,Fv],['lightgreen','green'])
plt.xlim(-4,4)
plt.ylim(-1,5)
```

Applying affine transformations to a vector

```python
A = np.array([[-1,4],[2,-2]])
```

```python
Av = np.dot(A,x)
plot_vectors([x,Av],['lightblue','blue'])
plt.xlim(-1,5)
plt.ylim(-1,5)
```

```python
#another example of applying A
x2 = np.array([2,1])
Av2 = np.dot(A,x2)
plot_vectors([x2,Av2],['lightgreen','green'])
plt.xlim(-1,5)
plt.ylim(-1,5)
```

Define 4 vectors to visualize

```python
v1 = np.array([3,1])
v2 = np.array([2,1])
v3 = np.array([-3,-1])
v4 = np.array([-1,1])
```

To apply the A vector and see how the other vectors are linearly transformed

```python
V = np.concatenate([np.matrix(v1).T,np.matrix(v2).T,np.matrix(v3).T,np.matrix(v4).T],axis = 1)
Av = np.dot(A,V)
```

To extract the column vectors

```python
def vectorfy(mtrx,col):
    return np.array(mtrx[:,col]).reshape(-1)
```

```python
plot_vectors([vectorfy(V,i) for i in range(0,4)]+[vectorfy(Av,i) for i in range(0,4)],
             ['lightblue','lightgreen','pink','lightgrey',
              'blue','green','red','grey'])
plt.xlim(-4,6)
plt.ylim(-5,5)
```
